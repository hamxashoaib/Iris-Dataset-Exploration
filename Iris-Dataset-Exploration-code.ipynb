# Task 1: Exploring and Visualizing the Iris Dataset

## Problem Statement
The objective of this task is to understand how to load, inspect, and visualize a dataset to identify data trends and distributions. We will work with the famous Iris dataset, which contains measurements of iris flowers from three different species.

## Goal
- Load the Iris dataset using pandas
- Perform exploratory data analysis (EDA)
- Generate visualizations to understand feature relationships and distributions
- Extract meaningful insights from the data

---

## 1. Import Required Libraries
We'll import the necessary libraries for data manipulation, visualization, and analysis.

```python
# Data manipulation libraries
import pandas as pd
import numpy as np

# Data visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# For loading the Iris dataset
from sklearn.datasets import load_iris

# Set style for better-looking plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

print("✓ All libraries imported successfully!")
```

## 2. Load the Dataset
Load the Iris dataset using sklearn and convert it to a pandas DataFrame for easier manipulation.

```python
# Load the Iris dataset from sklearn
iris_data = load_iris()

# Create a DataFrame for easier manipulation
df = pd.DataFrame(
    data=iris_data.data,
    columns=iris_data.feature_names
)

# Add the target variable (species)
df['Species'] = iris_data.target_names[iris_data.target]

print("✓ Dataset loaded successfully!")
print(f"\nDataset shape: {df.shape}")
```

## 3. Dataset Inspection
Let's examine the structure and contents of the dataset.

### 3.1 Display Basic Information

```python
# Display dataset shape
print("Dataset Shape:")
print(f"  Rows: {df.shape[0]}")
print(f"  Columns: {df.shape[1]}")

# Display column names
print("\nColumn Names:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col}")

# Display data types
print("\nData Types:")
print(df.dtypes)
```

### 3.2 First Few Rows

```python
# Display the first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Display the last 5 rows
print("\nLast 5 rows of the dataset:")
print(df.tail())
```

### 3.3 Dataset Information and Missing Values

```python
# Display dataset info
print("Dataset Information:")
df.info()

# Check for missing values
print("\nMissing Values:")
missing_values = df.isnull().sum()
print(missing_values)
print(f"\n✓ No missing values found!" if missing_values.sum() == 0 else "⚠ Missing values detected!")
```

### 3.4 Statistical Summary

```python
# Display statistical summary
print("Statistical Summary of Numeric Features:")
print(df.describe())

# Display count of each species
print("\nSpecies Distribution:")
print(df['Species'].value_counts())
```

## 4. Data Visualization
Create various visualizations to understand relationships between features and data distributions.

### 4.1 Scatter Plot - Relationship Between Features

```python
# Create a scatter plot showing relationships between Sepal Length and Sepal Width
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter plot 1: Sepal Length vs Sepal Width
for species in df['Species'].unique():
    data = df[df['Species'] == species]
    axes[0].scatter(
        data['sepal length (cm)'],
        data['sepal width (cm)'],
        label=species,
        alpha=0.7,
        s=100,
        edgecolors='black',
        linewidth=0.5
    )

axes[0].set_xlabel('Sepal Length (cm)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Sepal Width (cm)', fontsize=12, fontweight='bold')
axes[0].set_title('Sepal Length vs Sepal Width by Species', fontsize=13, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Scatter plot 2: Petal Length vs Petal Width
for species in df['Species'].unique():
    data = df[df['Species'] == species]
    axes[1].scatter(
        data['petal length (cm)'],
        data['petal width (cm)'],
        label=species,
        alpha=0.7,
        s=100,
        edgecolors='black',
        linewidth=0.5
    )

axes[1].set_xlabel('Petal Length (cm)', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Petal Width (cm)', fontsize=12, fontweight='bold')
axes[1].set_title('Petal Length vs Petal Width by Species', fontsize=13, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('scatter_plots.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Scatter plots created and saved as 'scatter_plots.png'")
```

### 4.2 Distribution Plots - Feature Distributions by Species

```python
# Create distribution plots for each feature
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

for idx, feature in enumerate(features):
    for species in df['Species'].unique():
        data = df[df['Species'] == species][feature]
        axes[idx].hist(data, label=species, alpha=0.6, bins=15, edgecolor='black')
    
    axes[idx].set_xlabel(feature, fontsize=11, fontweight='bold')
    axes[idx].set_ylabel('Frequency', fontsize=11, fontweight='bold')
    axes[idx].set_title(f'Distribution of {feature}', fontsize=12, fontweight='bold')
    axes[idx].legend()
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('distribution_plots.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Distribution plots created and saved as 'distribution_plots.png'")
```

### 4.3 Pair Plot - Relationships Between All Features

```python
# Create a pair plot to show relationships between all features
fig = plt.figure(figsize=(12, 10))
pair_plot = sns.pairplot(df, hue='Species', diag_kind='hist', palette='husl', plot_kws={'edgecolor': 'black', 'alpha': 0.7})

pair_plot.fig.suptitle('Pair Plot: Relationships Between All Features', fontsize=16, fontweight='bold', y=1.00)
plt.savefig('pair_plot.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Pair plot created and saved as 'pair_plot.png'")
```

### 4.4 Box Plots - Feature Variation by Species

```python
# Create box plots to show feature variation by species
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, feature in enumerate(features):
    sns.boxplot(data=df, x='Species', y=feature, ax=axes[idx], palette='Set2')
    axes[idx].set_xlabel('Species', fontsize=11, fontweight='bold')
    axes[idx].set_ylabel(feature, fontsize=11, fontweight='bold')
    axes[idx].set_title(f'Box Plot: {feature}', fontsize=12, fontweight='bold')
    axes[idx].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('box_plots.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Box plots created and saved as 'box_plots.png'")
```

### 4.5 Correlation Heatmap - Feature Correlations

```python
# Create a correlation heatmap
fig, ax = plt.subplots(figsize=(10, 8))

# Calculate correlation matrix (excluding species column)
correlation_matrix = df.drop('Species', axis=1).corr()

# Create heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=1,
    cbar_kws={'label': 'Correlation Coefficient'},
    ax=ax
)

ax.set_title('Feature Correlation Heatmap', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Correlation heatmap created and saved as 'correlation_heatmap.png'")
```
